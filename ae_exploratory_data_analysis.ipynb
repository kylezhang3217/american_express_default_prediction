### DATA TRANSFORMATION

```python
cat_features = ["B_30","B_38","D_114","D_116","D_117","D_120","D_126","D_63","D_64","D_66","D_68"]
num_features = [col for col in data.columns if col not in cat_features]
```

```python
train_num_agg = data.groupby("customer_ID")[num_features].agg(['mean', 'std', 'min', 'max', 'last'])
train_num_agg.columns = ['_'.join(x) for x in train_num_agg.columns]
train_num_agg.reset_index(inplace = True)
```

```python
train_cat_agg = data.groupby("customer_ID")[cat_features].agg(['count', 'last', 'nunique'])
train_cat_agg.columns = ['_'.join(x) for x in train_cat_agg.columns]
train_cat_agg.reset_index(inplace = True)
```

```python
data_transformed = train_num_agg.merge(train_cat_agg, how = 'inner', on = 'customer_ID')
```

```python
data_transformed
```

```python
data_w_labels = data_transformed.merge(data_labels, how = 'inner', on='customer_ID')
```

```python
# transform D_64_last = -1 value to 'E' for empty 
def transform_neg_1_to_e(value):
    if value == '-1':
        return 'E'
    else:
        return value
dvs.df['D_64_last'] = dvs.df['D_64_last'].apply(transform_neg_1_to_e)
```

```python
# drop rows with empty values
data_w_labels = dvs.df
data_w_labels = data_w_labels.dropna(axis=0)
```

```python
# persist data_w_labels for future use
import pickle 

def save_object(python_object, file_name):
    file_object_name = file_name + '.obj'
    file_object = open(file_object_name, 'wb')
    pickle.dump(python_object, file_object)
    file_object.close()
    
save_object(data_w_labels, 'american_express_data_w_labels')
```
